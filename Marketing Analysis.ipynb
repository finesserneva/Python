```python
#Data Cleaning

import pandas as pd
```

```python
facebook = pd.read_csv("facebook.csv")
facebook.columns
```

```python
facebook = facebook.rename(columns = {"spend" : "cost"})
```

```python
facebook = facebook[[
    "date",
    "campaign_name",
    "impressions",
    "clicks",
    "cost",
    "conversions"
]]
```

```python
facebook["platform"] = "Facebook"
facebook.head(10)
```

```python
facebook.info()
```

```python
google = pd.read_csv("google.csv")
```

```python
google.columns
```

```python
google = google[[
    "date",
    "campaign_name",
    "impressions",
    "clicks",
    "cost",
    "conversions"
]]
```

```python
google["platform"] = "Google"
```

```python
google.head(10)
```

```python
google.info()
```

```python
tiktok = pd.read_csv("tiktok.csv")
tiktok.columns
```

```python
tiktok = tiktok[[
    "date",
    "campaign_name",
    "impressions",
    "clicks",
    "cost",
    "conversions"
]]
```

```python
tiktok["platform"] = "TikTok"
```

```python
tiktok.head(10)
```

```python
tiktok.info()
```

```python
social_ads = pd.concat([facebook, google, tiktok])
```

```python
social_ads.head(10)
```

```python
social_ads.shape
```

```python
social_ads["platform"].value_counts()
```

```python
#Analysis time! 
#KPIs for all platforms
total_spend = social_ads["cost"].sum()
total_conversions = social_ads["conversions"].sum()
total_clicks = social_ads["clicks"].sum()
total_impressions = social_ads["impressions"].sum()
```

```python
conversion_rate = total_conversions / total_clicks
cpa = total_spend / total_conversions
ctr = total_clicks / total_impressions
cpc = total_spend / total_clicks
```

```python
print("Total Cost:", round(total_spend, 2))
print("Total Conversions:", total_conversions)
print("Total Clicks:", total_clicks)
print("Total Impressions:", total_impressions)

print("\nConversion Rate:", round(conversion_rate, 4))
print("Cost per Conversion (CPA):", round(cpa, 2))
print("CTR:", round(ctr, 4))
print("CPC:", round(cpc, 2))
```
