```python
import pandas as pd

# -----------------------------
# Data Cleaning
# -----------------------------

# Rename column
amzn_sales.rename(columns={"Amount": "Amount in INR"}, inplace=True)

# Select relevant columns
columns_select = ["Date", "Category", "Qty", "Amount in INR", "Status", "ship-city"]
clean_table = amzn_sales[columns_select].copy()

# Drop rows missing critical values
clean_table = clean_table.dropna(subset=["Qty", "Amount in INR"])

# Clean city names
clean_table["ship-city"] = clean_table["ship-city"].str.strip().str.upper()

# Convert Date to datetime (FIXED FORMAT)
clean_table["Date"] = pd.to_datetime(clean_table["Date"], format="%m-%d-%y")

# Create Month column
clean_table["Month"] = clean_table["Date"].dt.to_period("M")

# Add USD Conversion
conversion_rate = 0.011

clean_table["Amount in USD"] = clean_table["Amount in INR"] * conversion_rate

# Filter valid (non-cancelled) sales
valid_sales = clean_table[clean_table["Status"] != "Cancelled"]
```

```python
# Question 1: What is the total revenue generated from completed (non-cancelled) orders?

total_revenue_INR = valid_sales["Amount in INR"].sum()

int(total_revenue_INR)
```

```python
# *in USD*

total_revenue_USD = valid_sales["Amount in USD"].sum()

int(total_revenue_USD)
```

```python
# Question 2: How much revenue does each product category generate?

revenue_by_category_INR = valid_sales.groupby("Category")["Amount in INR"].sum().astype(int)

revenue_by_category_INR.sort_values(ascending=False)
```

```python
# *in USD*

revenue_by_category_USD = valid_sales.groupby("Category")["Amount in USD"].sum().astype(int)

revenue_by_category_USD.sort_values(ascending=False)
```

```python
# Question 3: Which product category generates the highest revenue?

revenue_by_category_INR.sort_values(ascending=False).head(1)
```

```python
# *in USD*

revenue_by_category_USD.sort_values(ascending=False).head(1)
```

```python
# Question 4: Which product category has the highest cancellation rate?

cancellation_rate_pct = cancellation_rate * 100

top_cancellation = cancellation_rate_pct.sort_values(ascending=False).head(1)

round(top_cancellation, 2)
```

```python
# Question 5: Which cities generate the highest revenue (Top 5)?

top_5_cities = (valid_sales.groupby("ship-city")["Amount in INR"].sum().sort_values(ascending=False).head(5))

top_5_cities.astype(int)
```

```python
# *in USD*

top_5_cities = (valid_sales.groupby("ship-city")["Amount in USD"].sum().sort_values(ascending=False).head(5))

top_5_cities.astype(int)
```

```python
# Question 6: What percentage of total orders are cancelled?

total_orders = len(clean_table)

cancelled_orders = len(clean_table[clean_table["Status"] == "Cancelled"])

cancelled_percentage = (cancelled_orders / total_orders) * 100

round(cancelled_percentage, 2)
```

```python
# Question 7: How does revenue change month-to-month over time?

monthly_revenue = valid_sales.groupby("Month")["Amount in INR"].sum().sort_index()

monthly_revenue.plot(figsize=(10,5), title="Monthly Revenue")
```
